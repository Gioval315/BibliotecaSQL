-	Realizar un gráfico personalizado (aplicar un estilo atractivo) donde se pueda analizar las ventas de los 8 productos (Producto A, Producto B y Producto C, ingresar los que faltan) en tres ciudades (Bogotá, Cali, Bucaramanga).

```python
import matplotlib.pyplot as plt
import numpy as np

productos = ['Producto A', 'Producto B', 'Producto C']
Bogota = [50, 75, 100]
Cali = [200, 180, 150]
Bucaramanga = [250, 220, 300]

x = np.arange(len(productos))
width = 0.25

plt.bar(x - width, Bogota, width, label='Bogotá', color='blue')
plt.bar(x, Cali, width, label='Cali', color='green')
plt.bar(x + width, Bucaramanga, width, label='Bucaramanga', color='red')

plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title('Ventas por Producto')

plt.xticks(x, productos)

plt.legend()

plt.show()
```

-	Realizar un gráfico personalizado (aplicar un estilo atractivo) para visualizar la evolución de las ventas de tres productos (Papa, Cebolla, Tomate) a lo largo de 8 meses en tres ciudades diferentes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataf = pd.DataFrame ({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto'] * 3,
    'Ventas': [100, 150, 200, 250, 300, 400, 300, 150,  # Papa (Cali)
               90, 120, 160, 200, 240, 100, 200, 150,    # Cebolla (Bogotá)
               80, 110, 150, 190, 230, 330, 350, 240],   # Tomate (Cartagena)
    'Producto': ['Papa (Cali)'] * 8 + ['Cebolla (Bogotá)'] * 8 + ['Tomate (Cartagena)'] * 8})

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(data=dataf, x='Mes', y='Ventas', hue='Producto', style='Producto', markers=True, dashes=False)

plt.title('Evolución de Ventas de Productos en Diferentes Ciudades', fontsize=16, fontweight='bold')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

plt.xticks(rotation=45)
plt.legend(title='Producto', loc='upper left')

plt.tight_layout()
plt.show()
```

-	Realizar un gráfico personalizado (aplicar un estilo atractivo) para visualizar la distribución de las puntuaciones en un examen según el nivel de motivación de los estudiantes, utilizando histogramas separados para baja, media y alta motivación. (Honestamente aqui me ayudo muchisimo GPT)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Puntuación': [50, 60, 65, 70, 75, 78, 80, 82, 85, 87, 90, 92, 94, 96, 98, 99, 100, 102, 103, 105],
    'Nivel de motivación': ['Baja', 'Media', 'Media', 'Alta', 'Alta', 'Media', 'Media', 'Alta', 'Alta', 'Baja',
                            'Alta', 'Alta', 'Media', 'Alta', 'Alta', 'Media', 'Alta', 'Media', 'Baja', 'Alta']
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Puntuación", hue="Nivel de motivación", multiple="stack", palette="viridis", binwidth=5)

plt.title('Distribución de las Puntuaciones según el Nivel de Motivación', fontsize=16)
plt.xlabel('Puntuación en el Examen', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
          'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
ventas_suv = [120, 150, 180, 200, 190, 170, 160, 180, 210, 220, 230, 250]
ventas_sedan = [100, 120, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230]
ventas_hatchback = [80, 90, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200]

data = {'Mes': meses, 'SUV': ventas_suv, 'Sedan': ventas_sedan, 'Hatchback': ventas_hatchback}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['SUV'], label='SUV', marker='o')
plt.plot(df['Mes'], df['Sedan'], label='Sedan', marker='s')
plt.plot(df['Mes'], df['Hatchback'], label='Hatchback', marker='^')

plt.title('Ventas Mensuales de Vehículos')
plt.xlabel('Mes')
plt.ylabel('Número de Vehículos Vendidos')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)

plt.show()
```
